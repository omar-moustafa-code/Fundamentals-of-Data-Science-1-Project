```r
install.packages("DescTools")
install.packages("ggplot2")
```

```r
heart = read.csv ("https://raw.githubusercontent.com/ahmedmoustafa/Notebooks/main/misc/heart.csv", header = TRUE)
head(heart) # first six rows of the data set
```

```r
tail(heart, 2) # last two rows of the data set
```

```r
dim(heart) # dimensions; returns number of rows then columns
```

```r
summary(heart) # six number summary of the data set
```

```r
min(heart$age) # age of youngest participant(s)
max(heart$age) # age of oldest participant(s)
```

```r
ncol(heart) # total number of variables in the data set
```

```r
plot(heart$age, pch = 19, cex = 1, col = 1)
```

```r
hist(heart$age)
```

```r
p <- ggplot(heart, aes(age))
p + geom_histogram()
```

```r
p <- ggplot(heart, aes(x = chol, fill = as.factor(diagnosis)))
p + geom_histogram(alpha = 0.7)
```

```r
p <- ggplot(heart, aes(x = as.factor(diagnosis), y = chol, fill = as.factor(diagnosis)))
p + geom_boxplot(alpha = 0.7)
```

```r
p <- ggplot(heart, aes(x = as.factor(diagnosis), y = chol, fill = as.factor(sex)))
p + geom_boxplot(alpha = 0.7)
```

```r
p <- ggplot(heart, aes(x = age, y = chol, color = as.factor(diagnosis)))
p + geom_point(alpha = 0.7)
```

```r
p <- ggplot(heart, aes(x = as.factor(cp), y = trtbps, fill = as.factor(diagnosis)))
p + geom_boxplot(alpha = 0.7)
```

```r
p <- ggplot(heart, aes(x = as.factor(sex), fill = as.factor(diagnosis)))
p + geom_bar(alpha = 0.7, position = "dodge")
```

```r
f <- Freq(heart$sex)
pie(f$freq, labels = f$level, main = "Genders and Heart Disease")
```
